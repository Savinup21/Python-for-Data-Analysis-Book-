**4.2 Pseudorandom Number Generation**

```python
import numpy as np
samples = np.random.standard_normal(size=(4,4))
samples
```

While Python's random module can only generate one random value at a time, the Numpy Random module is much faster

```python
from random import normalvariate
```

```python
n = 1_000_000
```

```python
%timeit samples = [normalvariate(0,1) for _ in range(n)]
```

The numbers that are created by this are not random, but instead *pseudorandom*, which means that they are generated by definite mathematical procedures 

```python
rng = np.random.default_rng(seed=12345)
data = rng.standard_normal((2, 3))
```
